## Задание 1 
 + а) Какие подгруппы товаров наиболее часто покупают за все время продаж (минимум 4 группы)?
 + б) Какие подгруппы товаров наиболее часто покупают за последние два года (минимум 4 группы)?
 + в) Какие подгруппы товаров наиболее часто покупают за последний год(минимум 4 группы)?

При помощи SQL-запросов сгруппируем по подгруппе и посчитаем кол-во заказов за разные периоды времени 

```python
sub_whole_time = spark.sql(''' SELECT `Sub-Category`, COUNT(ID) as total_orders
              FROM df
              GROUP BY `Sub-Category` 
              ORDER BY total_orders DESC
              LIMIT 5;''').toPandas()
sub_whole_time
```

```python
sub_two_years = spark.sql(''' SELECT `Sub-Category`, COUNT(ID) as total_orders
              FROM df
              WHERE EXTRACT(YEAR FROM `Order Date`) IN (2017 , 2018)
              GROUP BY `Sub-Category` 
              ORDER BY total_orders DESC
              LIMIT 5;''').toPandas()
sub_two_years
```

```python
sub_last_year = spark.sql(''' SELECT `Sub-Category`, COUNT(ID) as total_orders
              FROM df
              WHERE EXTRACT(YEAR FROM `Order Date`) IN (2018)
              GROUP BY `Sub-Category` 
              ORDER BY total_orders DESC
              LIMIT 5;''').toPandas()
sub_last_year
```

```python
heights = [sub_whole_time['total_orders'],sub_two_years['total_orders'],sub_last_year['total_orders']]
x = [sub_whole_time['Sub-Category'],sub_two_years['Sub-Category'],sub_last_year['Sub-Category']]
colors = ['Green','Red','Orange']
title=['все время' , 'последние два года', 'последний год']
fig,axes = plt.subplots(nrows = 1, ncols=3, figsize=(18,6))
for i in range(3):
    axes[i].bar(height = heights[i], x = x[i], color = colors[i], edgecolor = 'Black',linewidth = 2)
    axes[i].set_title(f'Самые популярные подгруппы за {title[i]}')
    axes[i].set_xlabel('Подгруппы')
    axes[i].set_ylabel('Кол-во заказов')
plt.subplots_adjust(wspace=0.5)
```

**Вывод**:  
+ Папки (Binders) являются самой популярной подгруппой товаров на протяжении всех анализируемых периодов — за всё время, за последние два года и за последний год.
+ Бумага (Paper), мебель (Furnishings) и телефоны (Phones) стабильно входят в топ-4 самых покупаемых товаров.

## Задание 2
Построить boxplot («Ящик с усами») на основе продаж (Sales). Найти мажоритарную черту (т.е. избавиться от аномалий и представить четкую картину распределения величин).

```python
def boxplot (data,labels,height):
    fig,axes = plt.subplots(nrows = 1, ncols = len(labels),figsize=(15,8), 
                            gridspec_kw={'width_ratios': [1, 1, 2]})
    for i,label in enumerate(labels):
        res = [data[data[label] == row][height].values for row in data[label].unique()]
        axes[i].boxplot(x = res,labels = data[label].unique(), vert=False)
        axes[i].set_title(f'Диаграмма размаха {height} от {label}')
        axes[i].set_ylabel(f'{label}')
        axes[i].set_xlabel('Продажи')
    plt.subplots_adjust(wspace=0.5)
    plt.show()
    
```

```python
labels=df.columns[2:5]
height=df.columns[-1]
boxplot(df,labels,height)
```

Поскольку исходный boxplot не даёт чёткого представления о выбросах, я приведу столбец 'Sales' к логарифмическому масштабу, чтобы сжать диапазон значений и выявить аномалии.

```python
df['Sales_log']=np.log1p(df['Sales'])
df.head()
```

```python
height=df.columns[-1]
boxplot(df,labels,height)
```

```python
def filtered_data(data,column,target):
    for row in data[column].unique():
        Q1 = df[target].quantile(0.25)
        Q3 = df[target].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        data = data[(data[target] > lower_bound)&(data[target] < upper_bound)]
    return data
```

Поскольку наблюдаются выбросы, устраню их с помощью метода межквартильного размаха (IQR), удаляя значения, выходящие за пределы межквартильного размаха.

```python
labels=df.columns[2:5]
height=df.columns[-1]
df = filtered_data(df,'Class',height)
df = filtered_data(df,'Region',height)
df = filtered_data(df,'Sub-Category',height)
```

```python
boxplot(df,labels,height)
```

## Задание 3 
Для этого задания необходимо разбить все покупки на энное количество групп “Sale_group” (Допустим маленькие продажи, средние и высокие) на основе Sales
Сгруппировать данные на основе региона и группы продаж (Region, Sale_group). Определить основные тенденции и паттерны. Выделить наиболее «прибыльную» группу.

```python
description=df['Sales'].describe()
description
```

```python
df_sales_grouped = spark.sql(f'''WITH sg as (SELECT *,
              CASE
                  WHEN Sales <= {description['25%']} THEN 'low_sale'
                  WHEN Sales >= {description['75%']} THEN 'high_sale'
                  ELSE 'medium_sale'
              END as sale_group
              FROM df )
              SELECT Region,sale_group, COUNT(ID) AS total_users,
              SUM(Sales) as total_sales
              FROM sg
              GROUP BY Region,sale_group
              ORDER BY total_sales DESC''').toPandas()
```

```python
df_sales_grouped ['total_sales'] = df_sales_grouped ['total_sales'] / 1e6
df_sales_grouped
```

```python
colors = ['Pink','Green','Red','Orange']

fig,axes = plt.subplots(ncols = 4, nrows = 1,figsize = (30,8))
for i,row in enumerate(df_sales_grouped['Region'].unique()):
    res = df_sales_grouped[df_sales_grouped['Region'] ==row ]
    axes[i].bar(height = res['total_sales'],x = res['sale_group'],edgecolor = 'Black',linewidth = 2,color=colors[i])
    axes[i].set_title(f'Диаграмма продаж для региона {row}')
    axes[i].set_xlabel('Группа продажи')
    axes[i].set_ylabel('Сумма продаж (млн руб)')
plt.subplots_adjust(wspace=0.5)
```
